```python
columns = df.columns  #clean data
for col in columns:   #there is an 'O' char instead of 0 (zero) in last row, and more such issues.
    df[col]=  df[col].apply(lambda x: int(x.lower().replace('o', '0').replace('i','1').replace('$','%').replace('%','')) if type(x) == str else x)
```

```python
for col in columns:
    df[col] = df[col].replace(np.nan, df[col].mean())
```

```python
#row 12 seems to have an erratic value.
df.loc[12,'Median house price'] = 730000
df.loc[6,'Joe Bloggs: 2 year forecast'] = 15
```

```python
df['Actual Change'] = df['Median house price'].pct_change() * 100
```
